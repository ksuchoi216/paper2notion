```python
import requests
import os

load_dotenv()

# 환경변수 또는 직접 입력
NOTION_API_KEY = os.getenv("NOTION_API_KEY")  # 노션 Integration API 키
DATABASE_ID = os.getenv("NOTION_DATABASE_ID")  # 데이터베이스 ID
NOTION_VERSION = "2022-06-28"
# 환경변수 체크
if not NOTION_API_KEY or not DATABASE_ID:
    raise ValueError(
        "❌ NOTION_API_KEY 또는 NOTION_DATABASE_ID 환경변수가 비어있습니다."
    )

# 추가할 페이지의 데이터
new_page = {
    "parent": {"database_id": DATABASE_ID},
    "properties": {
        "Name": {"title": [{"text": {"content": "Third Entry"}}]},
        "Genre": {"rich_text": [{"text": {"content": "option4"}}]},
        "Completed": {"checkbox": True},
        "Number": {"number": 42},
    },
}

# API 요청
response = requests.post(
    "https://api.notion.com/v1/pages",
    headers={
        "Authorization": f"Bearer {NOTION_API_KEY}",
        "Notion-Version": NOTION_VERSION,
        "Content-Type": "application/json",
    },
    json=new_page,
)

# 결과 출력
if response.status_code == 200 or response.status_code == 201:
    print("✅ 페이지가 성공적으로 생성되었습니다!")
    print(response.json())
else:
    print("❌ 오류 발생:", response.status_code)
    print(response.text)
```

```python
import requests
import os
from dotenv import load_dotenv

# .env 로드
load_dotenv()

# 환경변수 또는 직접 입력
NOTION_API_KEY = os.getenv("NOTION_API_KEY")  # 노션 Integration API 키
DATABASE_ID = os.getenv("PAPER_DATABASE_ID")  # 데이터베이스 ID
NOTION_VERSION = "2022-06-28"

# 환경변수 체크
if not NOTION_API_KEY or not DATABASE_ID:
    raise ValueError(
        "❌ NOTION_API_KEY 또는 PAPER_DATABASE_ID 환경변수가 비어있습니다."
    )

# 추가할 페이지 데이터
new_page = {
    "parent": {"database_id": DATABASE_ID},
    "properties": {
        "Title": {  # ✅ Title은 반드시 있어야 함
            "title": [{"text": {"content": "Sample Paper Entry"}}]
        },
        "Year": {"number": 2025},  # ✅ 숫자형
        "Citation": {"number": 15},  # ✅ 숫자형
        "Summary": {  # ✅ rich_text
            "rich_text": [{"text": {"content": "This is a summary of the paper."}}]
        },
        "Tag": {"multi_select": [{"name": "AI"}, {"name": "ML"}]},  # ✅ multi_select
        "Done": {"checkbox": False},  # ✅ 체크박스
        "URL": {"url": "https://example.com"},  # ✅ URL
        "Finish": {"checkbox": False},  # ✅ 체크박스
        # "End": {"date": {"start": "2025-07-06"}},  # ✅ 날짜
        "End": {"date": None},  # ✅ 날짜
    },
}

# API 요청
response = requests.post(
    "https://api.notion.com/v1/pages",
    headers={
        "Authorization": f"Bearer {NOTION_API_KEY}",
        "Notion-Version": NOTION_VERSION,
        "Content-Type": "application/json",
    },
    json=new_page,
)

# 결과 출력
if response.status_code in [200, 201]:
    print("✅ 페이지가 성공적으로 생성되었습니다!")
    print(response.json())
else:
    print("❌ 오류 발생:", response.status_code)
    print(response.text)
```
